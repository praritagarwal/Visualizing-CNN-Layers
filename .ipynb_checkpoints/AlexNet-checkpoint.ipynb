A quick diversion to generate Figure 3 in [this](https://papers.nips.cc/paper/4824-imagenet-classification-with-deep-convolutional-neural-networks.pdf) paper. The CNN architecture proposed by them is now called AlexNet (for e.g. in [this](https://arxiv.org/abs/1312.6199) paper).

The pretrained model provided by pytorch is the one presented in [this](https://arxiv.org/pdf/1404.5997.pdf) paper which is  minor variation of the one in the first paper. 

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from torchvision import models
```

```python
AlexNet = models.alexnet(pretrained = True)
```

```python
AlexNet
```

```python
# extracting the first convolutional layer of AlexNet
KernelsFirstCNNLyr=AlexNet.features[0].weight.detach().clone()
```

```python
KernelsFirstCNNLyr.shape
```

```python
# function to 
# a) convert torch tensor to numpy
# b) make move the color channel to be the last axis
# c) renormalize the pixel values to lie between 0 and 1
def image_converter(image):
    im = image.numpy().transpose(1,2,0)
    immax = im.max()
    immin = im.min()
    im = (im - immin)/(immax - immin)
    return im
```

```python
nrows = 8
ncols = 8
ctr = 0


fig, ax = plt.subplots(figsize = (20,20), nrows = nrows, ncols = ncols )
for row in range(nrows):
    for col in range(ncols):
        ax[row, col].imshow(image_converter(KernelsFirstCNNLyr[ctr]))
        ax[row, col].set_xticks([])
        ax[row, col].set_yticks([])
        ax[row, col].set_title('kernel: {}'.format(ctr))
        ctr+=1                
plt.show()
```
